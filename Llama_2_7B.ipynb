```python
!CMAKE_ARGS="-DLLAMA_CUBLAS=on" FORCE_CMAKE=1 pip install llama-cpp-python==0.1.78 numpy==1.23.4 --force-reinstall --upgrade --no-cache-dir --verbose
!pip install huggingface_hub
!pip install llama-cpp-python==0.1.78
!pip install numpy==1.23.4

model_name_or_path = "TheBloke/Llama-2-13B-chat-GGML"
model_basename = "llama-2-13b-chat.ggmlv3.q5_1.bin"
# the model is in bin format
from huggingface_hub import hf_hub_download
from llama_cpp import Llama

model_path = hf_hub_download(repo_id=model_name_or_path, filename=model_basename)



# GPU
lcpp_llm = None
lcpp_llm = Llama(
      model_path=model_path,
      n_threads=2, # CPU cores
      n_batch=512, # Should be between 1 and n_ctx, consider the amount of VRAM in your GPU.
      n_gpu_layers=32 # Change this value based on your model and your GPU VRAM pool.
      )
# See the number of layers in GPU
lcpp_llm.params.n_gpu_layers


prompt = " What are the pros/cons of ChatGPT vs Open Source LLMs?"
prompt_template=f'''SYSTEM: You're Llama-2 Model
USER: {prompt}
ASSISTANT:
'''


response=lcpp_llm(prompt=prompt_template, max_tokens=886, temperature=0.5, top_p=0.95, repeat_penalty=1.2, top_k=150, echo=True)

# print(response)
print(response["choices"][0]["text"])



```

```python
def format_prompt(prompt, system_prompt=""):
    if system_prompt.strip():
        return f"[INST] {system_prompt} {prompt} [/INST]"
    return f"[INST] {prompt} [/INST]"


SYSTEM_PROMPT = """
You're a Llama-2 Model AI
""".strip()

```

```python
prompt = """
Write a function in python that calculates the square of a sum of two numbers.
""".strip()

result = lcpp_llm(format_prompt(prompt, SYSTEM_PROMPT))
print(result["choices"][0]["text"])
```

```python
#Code Generation
prompt = """
Write a function in python that splits a list into 3 equal parts and returns a list
with a random element of each sublist.
""".strip()

result = lcpp_llm(format_prompt(prompt, SYSTEM_PROMPT))
print(result["choices"][0]["text"])
```
